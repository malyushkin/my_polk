```python
import numpy as np
import pandas as pd
```

```python
filename = 'my_polk_mln__upd'
prefix = 'my_polk_mln__upd__'
```

```python
df_ = pd.read_csv(f"../../data/{filename}.csv", engine='python', index_col=0)
```

```python
df_.head()
```

```python
df_.tags = df_.tags.apply(lambda x: x.strip('[]').replace('\'', '').split(', ') if isinstance(x, str) and len(x) > 0 else np.nan)
```

```python
df = df_[:]
```

```python
df.published = pd.to_datetime(df.published)
```

# –ê–Ω–∞–ª–∏–∑ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π

## 1. –°–∞–º—ã–µ –∞–∫—Ç–∏–≤–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ (–æ–±—â–∏–π)

```python
# –¢–æ–ø 100 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π. 1-–π –∑–∞–ø–æ—Å—Ç–∏–ª 6086 –ø–æ—Å—Ç–æ–≤ –ø–æ —Ç–µ–º–µ, 100-–π ‚Äì 188 –ø–æ —Ç–µ–º–µ.
top_num = 100
df.owner_id.value_counts().nlargest(top_num).head()
```

–í—ã–¥–µ–ª–∏–º –ø–æ –æ–¥–Ω–æ–º—É –ø–æ—Å—Ç—É –∫–∞–∂–¥–æ–≥–æ –∞–∫—Ç–∏–≤–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.

```python
top_users_df = df[df.owner_id.isin(df.owner_id.value_counts().nlargest(top_num).index)].groupby('owner_id').first()[['post_url', 'query', 'tags']]
```

```python
top_users_df.head()
```

```python
top_users_df = top_users_df.merge(df.owner_id.value_counts().nlargest(top_num), left_index=True, right_index=True)
```

```python
top_users_df.rename(columns={'owner_id': 'count'}, inplace=True)
```

```python
top_users_df.sort_values('count', ascending=False).to_csv(f"../../export/{prefix}top_users.csv")
```

## 2. –°–∞–º—ã–µ –∞–∫—Ç–∏–≤–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ (–ø–æ –º–µ—Å—è—Ü–∞–º ‚Äì –¥–µ–∫–∞–±—Ä—å-—è–Ω–≤–∞—Ä—å, –∞–ø—Ä–µ–ª—å-–º–∞–π)

```python
date_selected_df = df[(df.published.dt.month == 1) | 
                      (df.published.dt.month == 4) | 
                      (df.published.dt.month == 5) | 
                      (df.published.dt.month == 12)]
```

```python
date_selected_df = date_selected_df.assign(month=date_selected_df.published.dt.month)
date_selected_df = date_selected_df.assign(year=date_selected_df.published.dt.year); date_selected_df.head()
```

```python
date_selected_df = date_selected_df[date_selected_df.year >= 2017][['owner_id', 'month', 'year']]; date_selected_df.head()
```

```python
date_selected_df_max = pd.DataFrame(columns=['index', 'count', 'year', 'month'])

for year in date_selected_df.year.unique():
    for month in date_selected_df.month.unique():
        date_selected_df_max = date_selected_df_max.append(
            date_selected_df[(date_selected_df.year == year) & (date_selected_df.month == month)] \
                .owner_id.value_counts() \
                .nlargest(5) \
                .reset_index() \
                .rename(columns={'owner_id': 'count'}) \
                .assign(year=year) \
                .assign(month=month)
        )
```

```python
date_selected_df_max = date_selected_df_max[['year', 'month', 'index', 'count']].sort_values(['year', 'month'])
date_selected_df_max.reset_index(drop=True, inplace=True)
date_selected_df_max
```

```python
date_selected_df_max['sample'] = ''
```

```python
for idx, row in date_selected_df_max.iterrows():
    date_selected_df_max.at[idx, 'sample'] = df.loc[df.owner_id == row['index'], 'post_url'].values[0]
```

```python
date_selected_df_max.head()
```

```python
date_selected_df_max.to_csv(f"../../export/{prefix}top_users_by_seasons.csv")
```
